# Seaborn
Seaborn is a Python data visualization library built on top of Matplotlib.

Why itâ€™s useful: It provides a high-level interface for creating beautiful and informative statistical graphics with less code.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_excel("C:/Users/Shubham/Downloads/Excel course/datasets/ESD.xlsx")
print(data)
```

## Line Plot

* Compared `Age` across `Business` Unit with `Ethnicity` as hue.
* Learned how to match color `palettes` dynamically with the number of categories.

```python
num_categories = data["Ethnicity"].nunique()
color = sns.color_palette("viridis", num_categories)
sns.lineplot(data =data,x = "Business Unit", y = "Age", hue ="Ethnicity", palette= color )
plt.show()
```

```python
data = sns.load_dataset("tips")
print(data)
```

## FacetGrid with Barplot

* Split data by `smoker` status and compared `tip` by `day` with `sex` as hue.
* Learned about specifying `order` to control categorical axis.

```python
a = sns.FacetGrid(data,col = "smoker",height =2,hue= "sex")
day_order = data["day"].unique()
a.map(sns.barplot,"day","tip", order=day_order)
plt.show()
```

## Relational Plot (`relplot`)

* Explored relationship between `tip` and `total_bill`, with hue (`sex`), size (`smoker`), and column (`day`).
* Practiced combining multiple dimensions in a single visualization.

```python
sns.relplot(data= data,x ="tip",y ="total_bill",hue= "sex", col = "day",size = "smoker")
plt.show()
```

## Bar Plot with Custom Order and Palette

* Used barplot to compare `tip` across `days` with `sex` as hue.
* Customized using palette "`GnBu`" and manually set order (`Sun` â†’ `Sat` â†’ `Fri` â†’ `Thur`).

```python
sns.barplot(data =data,x = "day", y = "tip",hue = "sex",palette = "GnBu",order = ["Sun","Sat","Fri","Thur"])
plt.plot()
```

## Histogram + KDE

* Used Titanic dataset to visualize `age` distribution, grouped by `who`.
* Applied bins and kernel density estimation (`KDE`).

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8,5))
sns.histplot(
    data=data,
    x="age",
    hue="who",
    kde=True,
    bins=25,
    palette="viridis",   
    alpha=0.7,           
    edgecolor="black")

plt.title("Age Distribution by Passenger Type", fontsize=14, fontweight="bold")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

## Heatmap (Groupby Aggregation)

* Aggregated average `tip` by `day` and plotted with a heatmap.
* Learned about handling pandas `FutureWarning` with `observed`.

```python
gp = data.groupby("day", observed=True).agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

## Categorical Scatterplot (Tips Dataset)  
- Plotted `total bill` against `day`, split by `gender`.  
- Used `jitter` to reduce overlapping points.  
- Showed spending patterns across `weekdays` by `gender`.

```python
#stripplot
data = sns.load_dataset("tips")
sns.stripplot(data=data,x = "day",y = "total_bill" ,hue = "sex",dodge = True,jitter =0.2)
plt.show()
```

## Categorical Scatterplot (Tips Dataset)

## Stacked Density Plot (KDE with Fill)  
- Visualized distribution of `total bill` across `weekdays`.  
- Used filled `KDE` to show proportions by `day`.  
- Highlighted contribution of each day to overall distribution.  

```python
#kde plot
data = sns.load_dataset("tips")
sns.kdeplot(data = data,x = "total_bill",hue = "day",multiple ="fill")
plt.show()
```
